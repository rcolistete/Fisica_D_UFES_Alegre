# Equação de Onda Eletromagnética

## Equação de Onda (mecânica, eletromagnética, etc)

A equação de (propagação de) onda é uma ["Equação Diferencial Parcial" (EDP)](https://pt.wikipedia.org/wiki/Equa%C3%A7%C3%A3o_diferencial_parcial#:~:text=Uma%20equa%C3%A7%C3%A3o%20diferencial%20parcial%20ou,f%C3%ADsica%20matem%C3%A1tica%2C%20f%C3%ADsica%20e%20engenharia.), ou [PDE em inglês, Partial Differential Equation](https://mathworld.wolfram.com/PartialDifferentialEquation.html), de 2a ordem linear, com derivadas parciais segundas em relação às coordenadas espaciais (1, 2 e 3, para domínio 1D, 2D e 3D) e derivada segunda parcial em relação ao tempo, em que aparece um termo ao quadrado que é a velocidade $v$ de propagação da onda.

$$\nabla^2 f(\vec{r},t) = \frac{1}{v^2} \frac{\partial^2 f(\vec{r},t)}{\partial^2 t}$$

[Equação de Onda, Wave Equation - MathWorld](https://mathworld.wolfram.com/WaveEquation.html) :

### Equação de Onda em uma dimensão (espacial), 1D

A equação de (propagação de) onda em uma dimensão (espacial), 1D, é uma equação diferencial parcial com derivada parcial segunda em relação a uma coordenada espacial, por exemplo, em $x$ (ou outra coordenada espacial, como $r$) e derivada parcial segunda em relação ao tempo, em que aparece um termo ao quadrado que é a velocidade $v$ de propagação da onda.

$$\frac{\partial^2 f(x,t)}{\partial^2 x} = \frac{1}{v^2} \frac{\partial^2 f(x,t)}{\partial^2 t}$$

## Equação de Onda Eletromagética

Via equações de Maxwell (vide aula anterior), obtemos equações de onda para o campo elétrico $\vec{E}$, bem como para campo magnético $\vec{B}$. Para $\vec{E}$ :

$$\nabla^2 \vec{E}(\vec{r},t) = \frac{1}{v^2} \frac{\partial^2 \vec{E}(\vec{r},t)}{\partial^2 t}$$

### Laplaciano escalar e vetorial

O 1o termo da equação de onda para $\vec{E}$ ou $\vec{B}$ é o operador diferencial [Laplaciano vetorial](https://mathworld.wolfram.com/VectorLaplacian.html), uma generalização do operador diferencial [Laplaciano usual que é aplicado a funções escalares](https://mathworld.wolfram.com/Laplacian.html). Vejamos cada um.

#### Laplaciano escalar

Laplaciano escalar é um operador diferencial aplicado em função escalar, resultando em um escalar. Em 3 dimensões e coordenadas cartesianas, é bem simples :

$$\nabla^2 f(x,y,z) = \frac{\partial^2 f(x,y,z)}{\partial^2 x} + \frac{\partial^2 f(x,y,z)}{\partial^2 y} + \frac{\partial^2 f(x,y,z)}{\partial^2 z}$$

[Laplaciano, Laplacian - MathWorld](https://mathworld.wolfram.com/Laplacian.html) :

#### Laplaciano vetorial

Laplaciano vetorial é um operador diferencial aplicado em função vetorial, resultando em um vetor.  
A expressão formal é meio complicada, i. e., gradiente de divergente da função vetorial, tudo menos rotacional do rotacional da função vetorial :

$$\nabla^2 \vec{F} = \vec{\nabla} (\vec{\nabla} \cdot \vec{F}) - \vec{\nabla} \times (\vec{\nabla} \times \vec{F}) $$

Mas o resultado é simples em coordenadas cartesianas, basicamente é o Laplaciano escalar aplicado em cada componente da função vetorial :

$$\nabla^2 \vec{F}(x,y,z) = \nabla^2 F_x(x,y,z)\,\hat{i} + \nabla^2 F_y(x,y,z)\,\hat{j} + \nabla^2 F_z(x,y,z)\,\hat{k}$$

[Laplaciano vetorial, Vector Laplacian - MathWorld](https://mathworld.wolfram.com/VectorLaplacian.html) :

Para ver o Laplaciano vetorial em coordenadas cartesianas, vide ["Vector Laplacian - HandWiki"](https://handwiki.org/wiki/Vector_Laplacian) :

e em coord. cilíndricas e esféricas, vide ["Del in cylindrical and spherical coordinates - HandWiki"](https://handwiki.org/wiki/Del_in_cylindrical_and_spherical_coordinates) :

### Equações de Onda Eletromagnética para cada componente

Ou seja, usando a expressão de Laplaciano vetorial em coordenadas cartesianas, da expressão anterior de equação de onda para o vetor campo elétrico $\vec{E}$ :

$$\nabla^2 \vec{E}(\vec{r},t) = \frac{1}{v^2} \frac{\partial^2 \vec{E}(\vec{r},t)}{\partial^2 t}$$

obtemos 3 equações de onda para cada componente $E_x$, $E_y$, $E_z$ do vetor campo elétrico :

$$\nabla^2 E_x(x,y,z,t) = \frac{1}{v^2} \frac{\partial^2 E_x(x,y,z,t)}{\partial^2 t}$$

$$\nabla^2 E_y(x,y,z,t) = \frac{1}{v^2} \frac{\partial^2 E_y(x,y,z,t)}{\partial^2 t}$$

$$\nabla^2 E_z(x,y,z,t) = \frac{1}{v^2} \frac{\partial^2 E_z(x,y,z,t)}{\partial^2 t}$$

### Equação de Onda Eletromagética para $\vec{E} = E_x\, \hat{i}$

Escolhendo $\vec{E} = E_x\, \hat{i}$ (rotaciona-se o sistema de coordenadas para o eixo $x$ coincidir com o eixo de oscilação do campo elétrico $\vec{E}$), então só resta uma equação de onda para $\vec{E}$ :

$$\nabla^2 E_x(x,y,z,t) = \frac{1}{v^2} \frac{\partial^2 E_x(x,y,z,t)}{\partial^2 t}$$

E escolhendo o eixo de propagação da onda como $z$, então $E_x = E_x(z, t)$. Lembrando também que para a luz $v = c$. Logo a equação de onda para $\vec{E}$ fica bem simplificada

$$\frac{\partial^2 E_x(z,t)}{\partial^2 z} = \frac{1}{c^2} \frac{\partial^2 E_x(z,t)}{\partial^2 t}$$

### Equação de Onda Eletromagética para $\vec{B} = B_y\, \hat{j}$

Devido à escolha anterior de $\vec{E}$, então $\vec{B} = B_y\, \hat{j}$, então só resta uma equação de onda para $\vec{B}$ :

$$\nabla^2 B_y(x,y,z,t) = \frac{1}{v^2} \frac{\partial^2 B_y(x,y,z,t)}{\partial^2 t}$$

E escolhendo o eixo de propagação da onda como $z$, então $B_y = B_y(z, t)$. Lembrando também que para a luz $v = c$. Logo a equação de onda para $\vec{B}$ fica bem simplificada

$$\frac{\partial^2 B_y(z,t)}{\partial^2 z} = \frac{1}{c^2} \frac{\partial^2 B_y(z,t)}{\partial^2 t}$$

# Propagação da Luz/Onda Eletromagnética

Dadas as equações de onda eletromagnética para $E_x(z,t)$ e $B_y(z,t)$, as soluções podem ser do tipo senoidal progressiva :

$$\vec{E}(z,t) = E_x(z,t)\, \hat{i} = E_0 \sin(\kappa\,z - \omega\,t)\, \hat{i}$$

$$\vec{B}(z,t) = B_y(z,t)\, \hat{j} = B_0 \sin(\kappa\,z - \omega\,t)\, \hat{j}$$

Ou seja, $\vec{v} \propto \hat{k}$, $\vec{E} \perp \vec{v}$, $\vec{B} \perp \vec{v}$ e $\vec{E} \perp \vec{B}$, tal que $\vec{E} \times \vec{B} \propto \vec{v}$.

Além disso, substituindo nas equações de Maxwell, temos que :

$$B_0 = \frac{E_0}{c}$$

$$c = \frac{1}{\sqrt{\mu_0\,\epsilon_0}}$$
